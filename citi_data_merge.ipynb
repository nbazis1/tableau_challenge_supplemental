```python
import pandas as pd
```

```python
#Read all CSVs
jan22 = pd.read_csv('jan22.csv')
feb22 = pd.read_csv('feb22.csv')
mar22 = pd.read_csv('mar22.csv')
jun22 = pd.read_csv('jun22.csv')
jul22 = pd.read_csv('jul22.csv')
aug22 = pd.read_csv('aug22.csv')
```

```python
#Make winter & summer dataframes
winter22 = pd.concat([jan22, feb22, mar22])
summer22 = pd.concat([jun22, jul22, aug22])
```

```python
#View dataframes
winter22
summer22
```

```python
#Add "season" column, concatenate CSVs 
# Add a 'season' column to each DataFrame
winter22['season'] = 'winter'
summer22['season'] = 'summer'
winter_summer22 = pd.concat([winter22, summer22])
winter_summer22.to_csv('winter_summer22.csv', index=False)
```
